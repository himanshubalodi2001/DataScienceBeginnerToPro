```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
flights=sns.load_dataset("flights")
print(flights)
```

```python
f_sum=flights.groupby("month",observed=False)["passengers"].sum()
sns.boxplot(x="month",y="passengers",data=flights)
```

```python
sns.barplot(x="year",y="passengers",data=flights)
```

```python

import matplotlib.pyplot as plt
df = flights.groupby("year")["passengers"].sum()
plt.figure(figsize=(10, 6))
plt.plot(df.index, df.values,marker="o", linestyle='-', color='green')
plt.title("Total Passengers per Year (1949–1960)")
plt.xlabel("Year")
plt.ylabel("Total Passengers")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
df=flights.groupby("month",observed=False)["passengers"].sum()

plt.plot(df.index,df.values,marker='o',linestyle=":")
plt.title("TOTAL PASSENGERS BY MONTH")
plt.xlabel("Average Month")
plt.ylabel("Total Passengers")
plt.grid(True)
plt.tight_layout()
plt.show()
```
