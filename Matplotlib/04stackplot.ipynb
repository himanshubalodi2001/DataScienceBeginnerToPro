```python
import matplotlib.pyplot as plt
```

```python
activity=["study","job","sleep","gym"]
time_s=[4,8,9,2]
explode=[0.1,0,0,0]
plt.pie(time_s,labels=activity,autopct="%1.1f%%",shadow=True,explode=explode,wedgeprops={'edgecolor':"black","linestyle":'-'})
plt.legend()
plt.title("Time distributed")
plt.show()
```

```python
#stack plot
week_days=[0,1,2,3,4,5,6]
study=[4,5,2,3,6,4,3]
job=[8,8,8,8,8,8,8]
sleep=[8,9,6,7,9,5,6]
gym=[1,2,0,1,0,3,1]
labels=["study","job","sleep","gym"]
```

```python
plt.stackplot(week_days,study,job,sleep,gym,labels=labels,alpha=0.8,)
plt.legend()
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Weekly activity tracker")
plt.tight_layout()
plt.grid(True)
plt.show()
```
