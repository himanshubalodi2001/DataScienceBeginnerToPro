```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
tips=sns.load_dataset("tips")
tips.head(100)
```

```python
tips.describe()
```

```python
#average bill and average tip
#dont convert float to int after calculating then convert into int then round off
avg_bill=tips['total_bill'].mean()
print(f"average bill is = {round(avg_bill)}")
avg_tips=tips['tip'].mean()
print(f"average tips ={round(avg_tips)}")
```

```python
#group gender and smoker
a=tips.groupby('sex',observed=False)['smoker'].count()
print(a)
```

```python
#group male and time
b=tips.groupby(['time','sex'],observed=False)['sex'].count()
print(b)
```

```python
##group gender and bill
c=tips.groupby('sex',observed=False)['total_bill'].mean()
print(f"so the total bill given by male and female = {round(c)}")
```

```python
#group gender and bill
d=tips.groupby(['day','sex'],observed=False)['sex'].count()
print(d)
```

```python
tip=sns.load_dataset('tips')
sns.lineplot(x='total_bill',y='tip',data=tip)

plt.figure(figsize=(8, 5))
plt.scatter(tips['total_bill'], tips['tip'], alpha=0.7)
plt.xlabel('Total Bill')

plt.ylabel('Tip')
plt.title('Tips vs Total Bill')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.bar(tips['sex'], tips['tip'])
```

```python
df = tips.groupby(['day'], observed=False)['sex'].count()
plt.plot(df.index, df.values, marker='o')
plt.xlabel('Day')
plt.ylabel('Number of Customers')
plt.title('Number of Customers by Day')
plt.grid(True)
plt.show()
```
