```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.bar(years,runs,color="green")
plt.xlabel("years 1990-2010")
plt.ylabel("RUNS ")
plt.title("BAR graph example")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
x=np.arange(len(years))
width=0.25
plt.bar(x-width,sachin,width=width,label="Sachin T.")
plt.bar(x,sehwag,width=width,label="V. sehwag")
plt.bar(x+width,kohli,width=width,label="V. kohli")
plt.xlabel("YEAR 1990-2010")
plt.ylabel("RUNS")
plt.title("SACHIN-SEHWAG-Kohli comparison ")
plt.xticks(x,years)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]
plt.bar(players,runs,color="orange")
for i in range(len(runs)):
    plt.text(i,runs[i]+10,str(runs[i]),ha="center")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.tight_layout()
plt.show()
    
```
