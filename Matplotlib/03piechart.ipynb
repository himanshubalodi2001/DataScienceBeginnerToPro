```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot") 
```

```python
#simple
labels=["sachin","Dhoni","Kohli","rohit"]
runs=[10000,8000,5000,6000]
colors=["RED","GREEN","Black","orange"]
plt.pie(runs,labels=labels,colors=colors)
plt.legend()
plt.show()
```

```python
#edges,linewidth and colors
labels=["sachin","Dhoni","Kohli","rohit"]
runs=[10000,8000,5000,6000]
colors=["RED","GREEN","Black","orange"]
plt.pie(runs,labels=labels,colors=colors,wedgeprops={'edgecolor':'Purple','linewidth':2,'linestyle':'--'})
plt.show()
```

```python
#pull out like a sllice
labels=["sachin","Dhoni","Kohli","rohit"]
runs=[10000,8000,5000,6000]
colors=["RED","GREEN","Black","orange"]
explode=[0.1,0,0,0]
plt.pie(runs,labels=labels,colors=colors,explode=explode)
plt.show()
```

```python
#shadow
labels=["sachin","Dhoni","Kohli","rohit"]
runs=[10000,8000,5000,6000]
colors=["RED","GREEN","Black","orange"]
explode=[0.1,0,0,0]
plt.pie(runs,
        labels=labels,
        colors=colors,
        explode=explode,
        shadow=True,  #shadow
        startangle=150 # angel
       )
plt.show()
```

```python
labels=["sachin","Dhoni","Kohli","rohit"]
runs=[10000,8000,5000,6000]
colors=["RED","GREEN","Black","orange"]
explode=[0.1,0,0,0]
plt.pie(runs,
        labels=labels,
        colors=colors,
        explode=explode,
        shadow=True,  #shadow
        startangle=150, # angel
        autopct='%1.1f%%' # %begins a format specifier , 1.1f format as float with only one decimal point,%% to escaped %
       )
plt.show()
```
